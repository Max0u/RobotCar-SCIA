```python
import matplotlib.pyplot as plt
import cv2
import scipy
import scipy.misc
import os
import numpy as np
import skimage

import md

from time import time

import preprocess

%matplotlib inline
```

```python
data_dir = 'preprocessing/records_rev/'
X = [f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir, f))]
imgs = [cv2.imread(os.path.join(data_dir, x)) for x in X]
imgs = [np.array(img[60:-20, :, :], copy=True) for img in imgs]
mod = md.build_model()
mod.load_weights('../car_repo/models/model-0,0YUV.h5')
```

```python
t = time()
resize = 0
for img in imgs:
    image = img.copy()
    copy = time() - t
    t = time()
    image = scipy.misc.imresize(image, (200, 66))
    resize = time() - t
    t = time()
    image = preprocess.rgb2yuv(image)
    yuv = time() - t

t = time()
imgs = [img for img, _ in map(preprocess.preprocess, imgs)]
pre = time() - t
t = time()
preds = mod.predict(np.array(imgs)).squeeze()
pred = time() - t

total = pre + pred
print('total: %f - %f' % (total, total / total))
print('resize: %f - %f' % (resize, resize / pre))
print('copy: %f - %f' % (copy, copy / pre))
print('yuv: %f - %f' % (yuv, yuv / pre))
print('pred: %f - %f' % (pred, pred / total))
print('prepro: %f - %f' % (pre, pre / total))
```
