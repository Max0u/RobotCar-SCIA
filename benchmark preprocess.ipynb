```python
import matplotlib.pyplot as plt
import cv2
import scipy
import scipy.misc
import os
import numpy as np
import skimage

from time import time

import preprocess

%matplotlib inline
```

```python
data_dir = 'preprocessing/records_rev/'
X = [f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir, f))]
imgs = [cv2.imread(os.path.join(data_dir, x)) for x in X]
imgs = [np.array(img[60:-20, :, :], copy=True) for img in imgs]
```

```python
t = time()
resize = 0
for img in imgs:
    image = img.copy()
    copy = time() - t
    t = time()
    image = scipy.misc.imresize(image, (200, 66))
    resize = time() - t
    t = time()
    image = preprocess.rgb2yuv(image)
    yuv = time() - t
total = resize + copy + yuv
print('total: %f - %f' % (total, total / total))
print('resize: %f - %f' % (resize, resize / total))
print('copy: %f - %f' % (copy, copy / total))
print('yuv: %f - %f' % (yuv, yuv / total))
```
