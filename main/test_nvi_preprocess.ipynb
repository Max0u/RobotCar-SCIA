```python
import cv2, os
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as mpplot
import nvi_utils
```

```python
file_path = "../resources/test_dataset/frame_54_gas_0.5_dir_-0.9048416528908471.jpg"
real_path = "../resources/2018_06_13_16_44/frame_1585_gas_0.0_dir_0.0.jpg"
image = mpimg.imread(real_path)
image = nvi_utils.resize(image)
steering_angle = 0
mpplot.imshow(image)
```

```python
res = nvi_utils.preprocess(image)
mpplot.imshow(res)
```

```python

res1, steering_angle = nvi_utils.random_flip(image, steering_angle)
mpplot.imshow(res1)
```

```python
res2, steering_angle = nvi_utils.random_translate(image, steering_angle, 100, 10)
mpplot.imshow(res2)
```

```python
res3 = nvi_utils.random_shadow(image)
mpplot.imshow(res3)
```

```python
res4 = nvi_utils.random_brightness(image)
mpplot.imshow(res4)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
hsv[:,:,2] = hsv[:,:,2] * 3
res5 = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
mpplot.imshow(res5) 
```

```python
res6 = cv2.cvtColor(res5, cv2.COLOR_RGB2GRAY)
res6 = cv2.cvtColor(res6, cv2.COLOR_GRAY2RGB)
mpplot.imshow(res6)
```

```python
res7 = np.array(255 * (res6 / 255)**0.65, dtype=np.uint8)
mpplot.imshow(res7)
```
