```python
import os
import md
import numpy as np
import matplotlib.pyplot as plt
import cv2
import preprocess
```

```python
data_dir = "road_simulator/src/0-0/"
```

```python
X = [f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir, f))]
imgs = [cv2.imread(os.path.join(data_dir, x)) for x in X]
angle = [float(f.split("_")[5].split(".j")[0]) for f in X]
speed = [float(f.split("_")[3]) for f in X]
```

Test dataset (10000 images)  

```python
plt.hist(angle, bins=50)
plt.show()
```

```python
mod = md.build_model()
mod.load_weights('../car_repo/models/model-0,0YUV.h5')
imgs = [img for img, _ in map(preprocess.preprocess, imgs)]
preds = mod.predict(np.array(imgs))
```

```python
difference = np.abs(preds - angle)
average_diff = np.mean(difference)

print(average_diff)

plt.title('Aprs')
plt.xlabel('Prediction')
plt.ylabel('Label')
plt.scatter(preds, angle, marker='.', alpha=0.5, c='orange')
plt.plot(np.linspace(-2, 2), np.linspace(-2, 2), c='mediumorchid')
plt.show()
```
